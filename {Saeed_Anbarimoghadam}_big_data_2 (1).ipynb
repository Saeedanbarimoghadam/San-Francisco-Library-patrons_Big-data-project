**T** (5pts): Implement `map` using iteration for lists/arrays

````python
def map(func, xs):

    result = []
    for x in xs:
        result.append(func(x))
    return result
map(lambda x: x*2, range(7))
````

**T** (5pts): Implement `filter` using iteration for lists/arrays

````python
def filter(func, xs):
    result = []
    for x in xs:
        if func(x):
            result.append(x)
    return result

filter(lambda x: x % 2 == 0, range(7))
````

**T** (5pts): Implement a function `flatten(xs: [[A]]): [A]` that converts a list of
lists into a list formed by the elements of these lists. For example:

```python
>>> a = [[1,2],[2,3],[3,[4]]]
>>> flatten(a)
[1,2,2,3,3,[4]]
```

````python
def flatten(xss):
    result = []
    for xs in xss:
        if isinstance(xs, list):
            result.extend(xs)
        else:
            result.append(xs)
    return result


a = [[1, 2, 3], [4, 5], [7, [8, 9]]]
result = flatten(a)
print(result)
````

**T** (5pts): Implement `reduceL` by reusing `reduceR`

````python
def reduceL(func, xs, init):
    result = init
    for x in reversed(xs):
        result = func(result, x)
    return result

def reduceR(func, xs, init):
    result = init
    for x in xs:
        result = func(result, x)
    return result


result_reduceL = reduceL(lambda x, y: x - y, range(7), 0)
result_reduceR = reduceR(lambda x, y: x - y, range(7), 0)

print("reduceL:", result_reduceL)
print("reduceR:", result_reduceR)
````

**T** (10pts): Implement `group_by` by reusing `reduceL`.

````python
def reduceL(func, xs, init):
    result = init
    for x in reversed(xs):
        result = func(result, x)
    return result

def group_by(classifier, xs):
    initial_dict = {}

    def reducer(acc, x):
        key = classifier(x)
        acc[key] = acc.get(key, []) + [x]
        return acc

    result = reduceL(reducer, xs, initial_dict)
    return result

result = group_by(lambda x: "even" if x % 2 == 0 else "odd", range(10))
print(result)
````

**T** (5pts): Implement `distinct` using `reduceL`. Also provide its function signature.

````python
# distinct(xs:[A]): [A]
from functools import reduce

def reduceL(func, xs, init):
    return reduce(func, reversed(xs), init)

def distinct(xs):
    initial_set = set()

    def reducer(acc, x):
        acc.add(x)
        return acc

    result = reduceL(reducer, xs, initial_set)
    return list(result)

a = [1, 2, 3, 1, 2, 3, 4, 5, 6, 5, 4, 3, 2, 1]
result = distinct(a)
print(result)
````

**T** (5pts): Implement `flatmap` and provide its function signature.

````python
# flatmap(func: A -> [B], [A]): [B]
def flatmap(func, xs):
    return [item for sublist in map(func, xs) for item in sublist]


result = flatmap(lambda x: list(range(x)), range(5))
print(result)
````

**T** (10pts): Implement a function `zip(xs: [A], ys: [B]): List[(A,B)]` that returns a list formed from this list and another list by combining the corresponding elements in pairs. If one of the two lists is longer
than the other, its remaining elements are ignored.

```python
>>> a = [1,2,3,4]
>>> b = [a,b,c,d,e]
>>> zip(a,b)
[(1, 'a'), (2, 'b'), (3, 'c'), (4,'d')]
```

````python
def zip_lists(xs, ys):
    return list(zip(xs, ys))
a = [2,3,4]
b = ['a','b','c','d']
result = zip_lists(a, b)
print(result)
````

**T** (10pts): Implement a function
`scanL(f: (acc: B, x: A) -> B, xs: [A], init: B) -> [B]`
that works like `reduceL` but instead of producing one final result, it also
returns all the intermediate results.

```python
>>> a = [2,3,4]
>>> scanL(a, 0, lambda x, y: x + y)
[0, 2, 5, 9]
```

````python
from functools import reduce

def scanL(func, xs, init):
    return [init := func(init, x) for x in xs]
result= scanL(lambda x, y: x + y, [2,3,4], 0)
print(result)
````

**T** (10pts) Implement a function that outputs the total amount of checkouts from members originally registered at a given location.

```python
>>> checkouts(patrons, "Noe Valley/Sally Brunn")
1355624
```

````python
def checkouts(xs, location):
    total_checkouts = sum(int(patron['Total Checkouts']) for patron in xs if patron['Home Library Definition'] == location)
    return total_checkouts

result = checkouts(patrons, "Mission")
print(result)
````

**T** (15pts) Implement a function that lists the number of renewals per location in a tuple. Example output for part of the dataset:

```python
>>> number_renewals(patrons)

[('Presidio', 431988),
 ('Mission', 1218976)]
```

````python
from collections import defaultdict

def number_renewals(xs):
    renewals_per_location = defaultdict(int)

    for patron in xs:
        location = patron['Home Library Definition']
        renewals = int(patron['Total Renewals'])
        renewals_per_location[location] += renewals

    return list(renewals_per_location.items())
result = number_renewals(patrons)
print(result)
````
